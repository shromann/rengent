# Set-up

1. Data Agent: Retrive all required data stock.
    + MCP:
        - getting data:
            - historical prices
            - news
            - alt data
2. Analyst Agnet (Multi-Modal): Analyses all data for a stock.
    + MCP:
        - making calculations.
        - sentiments
3. Decision Agent: Makes a trade or doesn't trade.
    + MCP:
        - making trades.

$$
\text{Data Agent} \rightarrow \text{Analyst Agent} \rightarrow \text{Decision Agent}
$$

```python
from langchain_ollama import ChatOllama
from langchain.agents import create_agent
from langchain.tools import tool
from gnews import GNews

from pydantic import BaseModel, Field

class NewsSearchInput(BaseModel):
    query: str = Field(description="The subject or topic to search news for.")

news = GNews(max_results=10)

@tool('news_search', args_schema=NewsSearchInput)
def get_news(query: str) -> list[dict]:
    """Search the web for news.
    It returs a list of dictionaries, where each dictionary has:
        'title': 
        'description': ,
        'published date': ,
        'url': 
        'publisher':
    """
    return news.get_news(key=query)


tools = [get_news]

# Initialize the model with your local Ollama instance
llm = ChatOllama(
    model="llama3.2",
    temperature=0,
)

agent = create_agent(llm, tools)
```

```python
input_message = {
    "role": "user",
    "content": "Recent NVIDIA news",
}

for step in agent.stream(
    {"messages": [input_message]},
    stream_mode="values",
):
    step["messages"][-1].pretty_print()
```
